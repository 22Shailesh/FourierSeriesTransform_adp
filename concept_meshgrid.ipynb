### this script explains how to make meshgrid and 3d plot of a function using numpy and python nested loops
https://numpy.org/doc/stable/reference/generated/numpy.meshgrid.html

numpy.meshgrid(*xi, copy=True, sparse=False, indexing='xy') \
Return coordinate matrices from coordinate vectors.

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### using numpy

```python
ns = 10
xs = np.arange(-3,3,1/ns)
ys = np.arange(-3,3,1/ns)
print("shape of xs",np.shape(xs), " and type of xs is ", type(xs))
#XG, YG = np.meshgrid(xs, ys) 
XG, YG = np.meshgrid(xs, ys, sparse=True) 
print("shape of XG", np.shape(XG), " and type of XG is ", type(XG))
print("shape of YG", np.shape(YG))
#print(YG)
ZG = XG**2 + YG**2
xxy = np.sin(ZG)
print("shape of ZG", np.shape(ZG))
```

```python
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize = [8,6])
ax = fig.gca(projection='3d')
surf = ax.plot_surface(XG, YG, xxy,cmap=cm.summer,linewidth=0)
fig.colorbar(surf,shrink=0.5,aspect=5)
plt.show()
```

#### similar plot using python nested for loops

```python
fig = plt.figure(figsize = [8,6])
plot1 = plt.axes(projection='3d')

for i in range(60):
    x = -3.0 + i*0.1
    for j in range(60):
        y = -3.0 + j*0.1
        #print(format(x, '.2f'),format(y, '.2f'),format(z, '.2f'))
        #print(type(y))     #float type
        z = x**2 + y**2
        plot1.scatter(x, y, z);
```
